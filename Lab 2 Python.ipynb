```python
import pandas as pd
import numpy as np

df = pd.read_csv('videos_data.csv', index_col=0, skiprows=0,delimiter=';')


df
```

## Here I've selected some 207 channels from Youtube and harvested their public data from the YouTube Data API v3.

1. Get the most viewed video and it's channel

```python
def answer_one():
    
    return df.loc[df['views'] == df['views'].max()][["video_title"]]
answer_one()
```

2. Return the most disliked video of most liked channel

```python
def answer_two():
    mlc = df.groupby("channel_title")['likes'].sum().nlargest(1).index[0]
    mdv = df.loc[(df.index == mlc)]
    mdv = mdv.loc[(mdv['dislikes'] == mdv['dislikes'].max())]
    return mlc,mdv['video_title'][0]
answer_two()
```

3. Get the least popular channel

```python
def answer_three():
    lpc = df.groupby("channel_title")['views'].sum().nsmallest(1)
    return lpc
answer_three()
```

4. Get the oldest and the youngest videos and the difference between it

```python
def answer_four():
    df['age_days'].idxmin()
    yv = df.loc[(df['age_days'] == df['age_days'].min())]
    ov = df.loc[(df['age_days'] == df['age_days'].max())]
    return ov['video_title'][0], yv['video_title'][0],ov['age_days'][0] - yv['age_days'][0]
answer_four()
```

5. Get top 10 channels with the highest user activity

```python
def answer_five():
    df['activity'] = df['likes']+df['dislikes']+df['comments']
    return df.groupby("channel_title")['activity'].sum().nlargest(1).index[0]
answer_five()
```
