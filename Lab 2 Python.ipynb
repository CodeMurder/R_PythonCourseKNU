```python
import pandas as pd
import numpy as np

df = pd.read_csv('Used_Car_Price_Data.csv', index_col=0, skiprows=0)

price=[]
milage = []
review_count = []

for i in df['Price($)']:
    x = i.replace(',','')
    p = int(x)
    price.append(p)

for i in df['Milage']:
    x = i.replace(',','')
    m = int(x)
    milage.append(m)
    
for i in df['Review_count']:
    x = i.replace(',','')
    r = int(x)
    review_count.append(r)
    


# Creating a new dataframe.
# I decide to drop, too, the feature 'Badge'. I don't know what is this

df = pd.DataFrame({'price':price, 'milage':milage,'review':df['Review'], 'review_count': review_count})
df
```

## This data set is collection of used car so you can choose it to build a machine learning model about prediction of prices and various other project. This data set consist of total 1000 rows

1. Get the most cheap and the most expensive car

```python
def answer_one():
    return df['price'].idxmin(),df['price'].idxmax()
answer_one()
```

2. Return all cars that have less than 90000 milage and the price less than or equals 10000

```python
def answer_two():
    return df.loc[(df['milage'] < 90000) & (df['price'] < 10000)]
answer_two()
```

3. Get all of most reviewed cars

```python
def answer_three():
    df1 = df.sort_values('review',ascending=False)
    return df1.loc[(df1['review'] == 5)]
answer_three()
```

4. Get the most cheap Land Cruiser

```python
def answer_four():
    df1 = df.filter(like = 'Land Cruiser', axis=0)
    return df1['price'].idxmin()
answer_four()
```

5. Get the best car of 2004 year production (based on milage and price)

```python
def answer_five():
    df1 = df.filter(like = '2004', axis=0)
    return df1.loc[(df1['milage']==df1['milage'].min()) | (df1['price']==df1['price'].min())]
answer_five()
```
